# Problem

We would like to analyze the operations of major airports around the country and recommend strategies to cut down on delays nationwide. The most important part of this task is understanding the characteristics and groupings of airports based on a dataset of departure and operational delays.

all of the percent columns are blue so we will take them out to find our y

```python
X = df.groupby(['airport'])[df.columns.drop(['year','percent on-time gate departures',
                                             'percent on-time airport departures',
                                             'percent on-time gate arrivals'])].mean()
```

```python
ss = StandardScaler()
ss.fit(X)
scaled = ss.transform(X)
```

```python
pca = PCA(random_state=42)
pca.fit(scaled)
transformed = pca.transform(scaled)
```

```python
for ind in range(1,8):
    print 'The first {} components explain {}% of the variance.'.format(
           ind, sum(pca.explained_variance_ratio_[:ind])*100)
```

```python
plt.scatter(transformed[:,0], transformed[:,1])
plt.title('PCA of Airport Delays and Cancellations')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
pca.explained_variance_ratio_[:2].sum()
```

```python
tsne = TSNE(random_state=42)
tsne = tsne.fit_transform(transformed)
```

```python
plt.scatter(tsne[:,0], tsne[:,1])
plt.title('TSNE of Airport Delays and Cancellations')
plt.xlabel('TSNE1')
plt.ylabel('TSNE2')
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(tsne)
cls = km.transform(tsne)
```

```python
silhouette_score(tsne, km.labels_, metric='euclidean')
```

```python
plt.scatter(tsne[:,0], tsne[:,1], c=km.labels_, cmap='plasma')
plt.title('TSNE of Airport Delays and Cancellations with KMean Clusters')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
Z = linkage(tsne, 'ward')
```

```python
c, coph_dists = cophenet(Z, pdist(tsne))
c
```

```python
plt.figure(figsize=(10,20))
plt.title('Dendrogram')
plt.xlabel('Distance')
cut2 = dendrogram(Z, labels=X.index, orientation='right', color_threshold=650, leaf_font_size=10)
```

```python
max_dist = 650
clusters = fcluster(Z, max_dist, criterion='distance')
clusters
```

```python
X['labels'] = clusters
```

```python
pca_bad = PCA(random_state=42)
second = pca_bad.fit_transform(X[(X['labels'] == 1) | (X['labels'] == 3)])
```

```python
for ind in range(1,4):
    print 'The first {} components explain {}% of the variance.'.format(
           ind, sum(pca_bad.explained_variance_ratio_[:ind])*100)
```

```python
plt.scatter(second[:,0], second[:,1])
plt.title('PCA of Bad Airports')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
from matplotlib.collections import LineCollection

from sklearn import manifold, cluster, covariance
from sklearn.metrics import euclidean_distances
```

```python
names = X.index
```

```python
X_true = scaled
```

```python
similarities = euclidean_distances(X_true)
```

```python
_, labels = cluster.affinity_propagation(similarities)
n_labels = labels.max()

for i in range(n_labels + 1):
    print('Cluster %i: %s' % ((i + 1), ', '.join(names[labels == i])))
```

```python
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=42,
                   dissimilarity="precomputed", n_jobs=1)
pos = mds.fit(similarities).embedding_

nmds = manifold.MDS(n_components=2, metric=False, max_iter=3000, eps=1e-12,
                    dissimilarity="precomputed", random_state=42, n_jobs=1,
                    n_init=1)
npos = nmds.fit_transform(similarities, init=pos)

# Rescale the data
pos *= np.sqrt((X_true ** 2).sum()) / np.sqrt((pos ** 2).sum())
npos *= np.sqrt((X_true ** 2).sum()) / np.sqrt((npos ** 2).sum())

# Rotate the data
clf = TSNE(random_state=42)
X_true = clf.fit_transform(X_true)

pos = clf.fit_transform(pos)

npos = clf.fit_transform(npos)

fig = plt.figure(1,facecolor='white')
ax = plt.axes([0., 0., 1., 1.])

s = 100
plt.scatter(X_true[:, 0], X_true[:, 1], c=labels, cmap='plasma', lw=0,
            label='True Position')

similarities = similarities.max() / similarities * 100
similarities[np.isinf(similarities)] = 0

# Plot the edges
start_idx, end_idx = np.where(pos)

segments = [[X_true[i, :], X_true[j, :]]
            for i in range(len(pos)) for j in range(len(pos))]
values = np.abs(similarities)
lc = LineCollection(segments,
                    zorder=0, cmap=plt.cm.Blues,
                    norm=plt.Normalize(0, values.max()))
lc.set_array(similarities.flatten())
lc.set_linewidths(0.5 * np.ones(len(segments)))
ax.add_collection(lc)

# Add a label to each node. The challenge here is that we want to
# position the labels to avoid overlap with other labels
for index, (name, label, (x, y)) in enumerate(
        zip(names, labels, X_true)):

    dx = x - X_true[0]
    dy = y - X_true[1]
    this_dx = dx[np.argmin(np.abs(dy))]
    this_dy = dy[np.argmin(np.abs(dx))]
    if this_dx > 0:
        horizontalalignment = 'left'
        x = x + .8
    else:
        horizontalalignment = 'right'
        x = x - .8
    if this_dy > 0:
        verticalalignment = 'bottom'
        y = y + .8
    else:
        verticalalignment = 'top'
        y = y - .8
    plt.text(x, y, name, size=8, 
             horizontalalignment=horizontalalignment,
             verticalalignment=verticalalignment,
             bbox=dict(facecolor='w',
                       edgecolor=plt.cm.spectral(label / float(n_labels)),
                       alpha=.6))

plt.show()
```
